```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
#split train and test dataset
def train_test(X,y,model,*,test_size=0.2,random_state=42):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=test_size,random_state=random_state)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    mse=mean_squared_error(y_test,y_pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y_test,y_pred)
    return mse,rmse,r2,model # Return mse, rmse, r2, and the fitted model
```

```python
#linear regression: 1 feature
labl="SalePrice"  # Corrected typo from "SalePrace" to "SalePrice"
feature="TotalBsmtSF" # Assumed "TotalBsmtSF" as a relevant single feature based on df.info(). "Total" does not exist.

# Select relevant columns and drop rows with any NaN values in those columns
df_for_model = df[[feature, labl]].dropna()

# Prepare X (features) and y (target) for the model
X = df_for_model[[feature]] # X should be a DataFrame for sklearn
y = df_for_model[labl]      # y can be a Series

linear_model=LinearRegression()

# Correct arguments for train_test function: X, y, and the model object
# Correct assignment for return values: train_test now returns mse, rmse, r2, and the fitted model.
linear_mse, linear_rmse, linear_r2, linear_fit = train_test(X, y, linear_model)

print("Linear RMSE:", linear_rmse)
print("Linear R2:", linear_r2)
```

```python
#slope and intercept
slope=linear_fit.coef_[0]
intercept=linear_fit.intercept_
print("Slope:",slope)
print("Intercept:",intercept)
```

```python
#Polynomial Regression
ploy_model=Pipeline([('poly',PolynomialFeatures(degree=2)),('linear',LinearRegression())])

poly_mse, poly_rmse,poly_r2,poly_fit=train_test(X,y,ploy_model)

print("Polynomial RMSE:",poly_rmse)
print("Polynomial R2:",poly_r2)
```
